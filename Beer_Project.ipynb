To help us wait, we wrote the following function which not only sleeps for a predetermined amount of time, but also lets us reload our own source code in the meantime (so we can take advantage of these large latency periods to update our scripts):

```python
# Sleep for a specified amount of seconds, and show the countdown
# @param secs Amount of seconds to sleep for.
def sleepCountdown(secs):
    while True:
        try:
            temp = secs
            print "Sleeping for %d seconds." % secs
            print("DAYS:HOURS:MIN:SEC")
            while(secs >= 0):
                sec = timedelta(seconds=secs)
                d = datetime(1,1,1) + sec
                sameLine("%02d:%02d:%02d:%02d" % (d.day-1, d.hour, d.minute, d.second))
                time.sleep(1)
                secs -= 1
            break
        
        # Allow us to reload this file on a keyboard interrupt
        except KeyboardInterrupt:
            print "\nInterrupt detected."
            print "Press ctrl+\ to exit"
            execfile("includes.py")
            print "Includes reloaded."
            continue
        
        
    print "\nSlept %d secsonds." % temp
```

We then stored the final data in a MySQL database, also running on the server. Since we storing our data in a MySQL database on an Amazon EC2 server, were able to leverage proper indexing and a large amount of computational power to quickly run complicated queries.

To access the MySQL database, we used the following utility functions:

```python
# return a Pandas data frame given a SQL query and optional parameters.
# @param query SQL query to execute
# @param arg1, arg2, arg3, etc. Parameters to passed into the SQL query (that replace %s parameters).
# @return Pandas data frame on success, None on failure
# @notes The SQL user used only has SELECT privelges on the MYSQL_DB database (for good reason).
def sqlDataFrame(query, *args):
    global MYSQL_HOST, MYSQL_USER, MYSQL_PASSWORD, MYSQL_DB, MYSQL_ERROR 
    df = None
    try:
        con = mdb.connect(MYSQL_HOST, MYSQL_SELECT_USER, MYSQL_SELECT_PASSWORD, MYSQL_DB);
        df = psql.read_sql(query, params=args, con=con)  

    # handle any exceptions
    except Exception as e:
        print traceback.format_exc()
        error = "MySQL Error %d: %s" % (e.args[0],e.args[1])
        print error
        MYSQL_ERROR = error
    finally:
        con.close()
        return df

# check the result of a MySQL query, emailing and quitting as necessary
# @param result Result of a query or sqlDataFrame function call.
def checkResult(result):
    global MYSQL_ERROR
    if result == None:
        sendEmail(MYSQL_ERROR)
        sys.exit(1)

# Exceutes a MySQL query given a SQL query and optional parameters.
# @param query SQL query to execute
# @param arg1, arg2, arg3, etc. Parameters to passed into the SQL query (that replace %s parameters).
# @return None on failure, the result of the query on success.
# @notes The SQL user used has all privileges on the MYSQL_DB database. USE WITH CAUTION.
def query(query, *args):
    global MYSQL_HOST, MYSQL_USER, MYSQL_PASSWORD, MYSQL_DB, MYSQL_ERROR, LAST_INSERT_ID
    result = None
    try:
        con = mdb.connect(MYSQL_HOST, MYSQL_USER, MYSQL_PASSWORD, MYSQL_DB);

        # handle unicode problems client side
        con.set_character_set('utf8')
        cur = con.cursor()
        
        # handle unicode errors server side
        cur.execute('SET NAMES utf8;') 
        cur.execute('SET CHARACTER SET utf8;')
        cur.execute('SET character_set_connection=utf8;')
         
        cur.execute(query, args)
        con.commit()
        LAST_INSERT_ID = cur.lastrowid

        result = cur.fetchall()

    except Exception as e:
        print traceback.format_exc()
        error = "MySQL Error %d: %s" % (e.args[0],e.args[1])
        print error
        MYSQL_ERROR = error
    
    finally:
        con.close()
        return result
```

First, we created the concept of a basic api request, here:

```python
# Generic Untappd API request
# @param method Name of API call
# @param parameters Optional paremters to send with the call
def apiRequest(method, params={}):
    global RATE_LIMIT, RATE_LIMIT_REMAINING
    print "Rate Limit Remaining: %d" % RATE_LIMIT_REMAINING
    
    if(RATE_LIMIT_REMAINING <= 1 and RATE_LIMIT_REMAINING != -1):
        sleepCountdown(ONE_HOUR)
    
    # add our client id and secret
    params["client_id"] = CLIENT_ID
    params["client_secret"] = CLIENT_SECRET
    response = requests.get(BASE_URL + method, params=params)
    
    try:
        json = response.json()
    except JSONDecodeError as err:
        raise(err)
    
    # make sure we got expected json back
    if not json or "meta" not in json or "code" not in json["meta"]:
        raise Exception("Error: %d: No additional info available" % response.status_code)

    # check the meta from the json
    if json["meta"]["code"] != 200:
        # handle weird 501s we're getting and sleep
        if json["meta"]["code"] == 501:
            print "Received 501 (something went wrong). Trying again in 10 seconds."
            sleepCountdown(10)
            return apiRequest(method, params)
            
        # handle a 500 and sleep for a whole hour
        if json["meta"]["code"] == 500:
            print "Received 500 (something's wrong on their end). Trying again in an hour."
            sleepCountdown(ONE_HOUR)
            return apiRequest(method, params)
        
        # handle an unknown exception
        RATE_LIMIT_REMAINING = 0
        raise Exception("Error %d (%s): %s" % (json["meta"]["code"], json["meta"]["error_detail"], json["meta"]["error_detail"]))
        
	# update global rate limit information
    RATE_LIMIT = int(response.headers["X-Ratelimit-Limit"])
    RATE_LIMIT_REMAINING = int(response.headers["X-Ratelimit-Remaining"])

    # we did it!
    return json
```

This allowed us to keep track of the global rate limit for any request, so we didn't exceed the 100 requests per hour the API allowed for.

### 2.3 Data Wrangling

At the same time that the previous script was running, we also ran [wrangleData.py](https://github.com/iannightingale/cs109/blob/master/py/wrangleData.py), copied below:

```python
#!/home/ubuntu/anaconda/bin/python
"""
wrangleData.py
Wrangles all the data fetched by getEverything.py, 
then sleeps for an hour to wait for more data to wrangle.
"""

# includes 
execfile("includes.py")
from geopy.distance import great_circle

# Geolocate every user
def step1():
    print "Updating..."
    results = query("SELECT id, username, location FROM users WHERE location <> '' AND lat IS NULL")
    i = 0

    for user_id, username, location in results:
        sameLine("Geolocating: %s (%d/%d)" % (username, i, len(results)))
        location = geolocator.geocode(location)
        if location == None:
            lat, lng = -1, -1
        else:
            lat, lng = location.latitude, location.longitude
        result = query("UPDATE users SET lat = %s, lng = %s WHERE id = %s", lat, lng, user_id)
        checkResult(result)
    
        i += 1

    print "\nFinished! step 1" 

# Calculate the distance for every rating
def step2():
    sql = """SELECT DISTINCT users.id, ratings.beer_id, users.lat, users.lng, breweries.lat, breweries.lng
        FROM ratings
        LEFT JOIN users ON ratings.user_id = users.id
        LEFT JOIN beers ON beers.id = beer_id
        LEFT JOIN breweries ON breweries.id = brewery_id
        WHERE users.lat <> -1 AND users.lat IS NOT NULL
        AND distance = -1
        AND user_rating > 0
        AND (breweries.lat <> 0 AND breweries.lng <> 0)
    """

    print "Querying..."
    results = query(sql)
    i = 0
    print "Updating..."
    for user_id, beer_id, lat1, lng1, lat2, lng2 in results:
        	sameLine("%d/%d" % (i, len(results)))
        	i += 1
        	km = great_circle((lat1, lng1), (lat2, lng2)).km
        	result = query("UPDATE ratings SET distance = %s WHERE user_id = %s AND beer_id = %s",
                       km, user_id, beer_id)
        	checkResult(result)
    print "\nFinished step 2!"

import re

# Tokenize every description
def step3():
    pattern1 = re.compile('[^\w\s]+')
    pattern2 = re.compile('\s+')
    beers = query("SELECT id, name, description FROM beers WHERE parsed = 0")
    checkResult(beers)
    i = 0
    print "Tokenizing.."
    for beer_id, name, description in beers:
        print("Tokenizing: %s (%d/%d)" % (name, i, len(beers)))

        tokens = pattern2.sub(' ', pattern1.sub('', description)).upper().strip().split(' ')

        for token in tokens:
            # watch for empty string case
            if token == "":
                continue
            
            word_ids = query("SELECT id FROM words WHERE word = %s", token)
            # insert the word
            if(len(word_ids) == 0):
                result = query("INSERT INTO words (word) VALUES(%s)", token)
                checkResult(result)
                word_id = LAST_INSERT_ID;
            else:
                word_id = word_ids[0][0]

            result = query("""INSERT INTO description_words (beer_id, word_id) VALUES(%s, %s) 
                ON DUPLICATE KEY UPDATE count = count + 1""", beer_id, word_id)

            checkResult(result)

        result = query("UPDATE beers SET parsed = 1 WHERE id = %s", beer_id)
        checkResult(result)

        i += 1
        
    # update Word Counts in Whitelist
    result = query("""UPDATE word_whitelist LEFT JOIN (SELECT word_id, sum(`count`) as cnt FROM description_words GROUP BY word_id) as tmp 
    ON word_whitelist.id = tmp.word_id
    SET word_whitelist.count = tmp.cnt
    WHERE tmp.word_id IS NOT NULL""")
    checkResult(result)
    
    print "\nFinished step 3!"

if __name__ == "__main__":
    try:
        # update superusers table with geo coordinates
        while True:
            step1()
            step2()
            step3()
            sleepCountdown(ONE_HOUR)
            
    except Exception as e:
           print "Error", sys.exc_info()[0]
           print traceback.format_exc()
           sendEmail("Error occurred in getLocations!")
```

This accomplished the following:
1. Found the latititude and longtitude for every user that hadn't been previously geolocated
2. Found the distance between the user and the beer for every user rating (using the location of the Brewery and the information from Step 1.
3. Tokenized the descriptions for every beer that hadn't previously been tokenized [(see below for explanation)](#languageAnalysis)

After finishing each of these three steps were completed, the script then always waited one hour. This ensured that their would be new data to wrangle, and maximized our effeciently by running the two scripts in parallel.

### 3.1 Style

The operational unit of beer is the style - a traditional (or sometimes not so traditional) characterization of a beer that defines its ingredients and flavor profile. ABV and IBUs are good post-production metrics of ingredients - ABV of grain used, IBU of hops - so we first set out to explore the relationship between these variables. We could first plot ABV vs IBU for some of the most frequent styles, noting how styles cluster in space:

```python
# first calculate the most frequent styles in our database for a data subset
beers = sqlDataFrame("""SELECT ratings.*, beers.* FROM `ratings` LEFT JOIN super_users ON (super_users.user_id = ratings.user_id) LEFT JOIN beers ON (beer_id = beers.id) 
                     WHERE ratings.user_rating > 0 AND super_users.user_id IS NOT NULL""")
# group by style and compute counts
counts = beers.groupby("style").count().sort("user_id", ascending=False)
counts["style"] = counts.index
# generate 20 beers to include in this plot
topStyles = list(counts.style[0:20])
print topStyles
```

```python
# define a color vector
colors = ['yellow', 'green','blue','red', "orange"]
include = {}
# make color dictionary
for style, color in zip(topStyles, colors):
    include[style] = color
    
# subselect beers for top styles
subset = df[df.style.isin(include.keys())]
# plot scatter
plt.scatter(subset.abv, subset.ibu, c=subset["style"].apply(lambda x: include[x]), s=25, alpha=0.6)

# plot legend
import matplotlib.patches as mpatches
plt.legend(handles=map(lambda x: mpatches.Patch(color=include[x], label=x), include.keys()))

plt.title("IBU vs. ABV for %d Most Frequent Beer Styles, n=%d" %(len(include), len(subset)))
plt.xlabel("ABV (%)")
plt.ylabel("IBU")
plt.xlim(0,25)
plt.ylim(0,200)
plt.show()
```

```python
# define a function to regress two quantitative variables and plot

import statsmodels.api as sm

# http://stackoverflow.com/questions/19068862/how-to-overplot-a-line-on-a-scatter-plot-in-python
def regplt(data, X,Y):
    x = np.array(data[X])
    y = np.array(data[Y])
    # build linear model
    results = sm.OLS(y,sm.add_constant(x)).fit()
    print results.summary()

    # plot scatter with regression line
    plt.scatter(x,y, alpha=0.4, c="black")
    X_plot = np.linspace(0,max(x),100)
    # plot line, transforming points by lin mod intercept and slope
    plt.plot(X_plot, X_plot*results.params[1] + results.params[0], color="red", linewidth=3)
    plt.title("LSQ Regression Line for %s vs. %s, n=%d" %(Y, X, len(data)))
    plt.xlabel(X)
    plt.ylabel(Y)
    plt.show()
```

### 5.2 Wrangling

We manually deleted words that were unrelated to beer or beer quality (words like "a," "is," "they", but also word like "beer" that are common to all beers in our database). To accomplish this, we first created a whitelist table that contained only the words we wanted, and used the following code:

```python
"""
returns the id of a word.
@param word Word.
"""
def getWordId(word):
    result = query("SELECT id FROM words WHERE word = %s", word)
    checkResult(result)
    return result[0][0]

"""
Adds the entries/counts for one set of words into another.

@param sources[] List of words to fold into a new word.
@param dest Destination to fold words into.
"""
def wordSwap(sources, dest):
    source_ids = map(getWordId, sources)
    destination_id = getWordId(dest)
    
    # iterate over soures
    for source_id in source_ids:
        # add old words into description word table
        result = query("""INSERT INTO description_words (beer_id, word_id, count)
            (SELECT beer_id, %s, count FROM description_words as tmp
            WHERE word_id = %s)
            ON DUPLICATE KEY UPDATE description_words.count = description_words.count + VALUES(count), 
            timestamp = CURRENT_TIMESTAMP""", destination_id, source_id)
        checkResult(result)
        # delete obsolete word from description words
        result = query("""DELETE FROM description_words WHERE word_id = %s""", source_id)
        checkResult(result)
        # delete obsolete word from the whitelist
        result = query("""DELETE FROM word_whitelist WHERE id = %s""", source_id)
        checkResult(result)
    
    # update the whitelist counts
    result = query("""UPDATE word_whitelist LEFT JOIN (
            SELECT word_id, SUM(  `count` ) AS cnt
            FROM description_words WHERE word_id = %s
            GROUP BY word_id
        ) AS tmp ON tmp.word_id = word_whitelist.id
        SET word_whitelist.count = tmp.cnt WHERE word_id = %s""", destination_id, destination_id)
    checkResult(result)
    
# example swap
wordSwap(["ZESTY"], "ZEST")
```

### 5.2 Description-based classifier

Given the results of this initial word analysis, we were interested in using the tokenized word counts in a predictive classifier - can these words help us to predict a user's rating of a beer, and if so, which ones are better than others? To do this, we processed each beer description in a particular user's beer list, counting the occurences of the top 1000 words found in the previous frequency analysis. We then implemented a random forest to generate predicted user rating based on the words contained in the description. The feature importance method allowed us to determine which words were most useful in our prediction algorithm.

```python
"""
Returns the occurences of the top words for a given user.

@param user_id User id to lookup.
@param limit Amount of words to use.
"""
def classifierDf(user_id, limit = 1000):
    combined = None
    # query for a non zero ratings for particular user
    results = query("""SELECT beer_id, user_rating FROM ratings WHERE user_rating > 0
                            AND user_id = %s""", user_id)
    i = 0
    # next query words for each beer
    for beer_id, user_rating in results:
        df = sqlDataFrame("""SELECT word, IFNULL( tmp.count, 0 ) 
                            FROM word_whitelist
                            LEFT JOIN (
                                SELECT * 
                                FROM description_words
                                WHERE description_words.beer_id = %s
                            ) AS tmp ON tmp.word_id = word_whitelist.id
                            ORDER BY word_whitelist.count DESC 
                            LIMIT %s""", beer_id, limit)
        
        df.columns = ["index", 0]
        df.loc[len(df)] = ("rating", user_rating)
        # create dataframe if None
        if combined is None:
            combined = df.T
            combined.columns = combined.ix[0,:]
            combined = combined.ix[1:]
        # or append to existing df
        else:
            combined.loc[len(combined)] = tuple(df.T.ix[1,:])
            
        # print progress in place
        sameLine("beer_id: %d (%d/%d)" % (beer_id, i, len(results)))
        i += 1
        
    return combined
```

The classifierDF function takes a user's id and creates a dataframe each of their beers as a row with the counts for each word in its description. We have an example for one user below.

## 8. Building a Basic Recommendation Tool

Finally, learning what we had above, we could implement our final goal - a recommendation tool that builds upon a random forest classifier to predict a user's rating of a new, unrated beer. **NB: User ratings were multiplied by 2 below to tranform them into integers.**

```python
from geopy.distance import great_circle

"""
@param user_id User id in our database.
@param beer Beginning of the beer name.
@param brewery Beginning of the brewery name.
@param k (Optional) If specified, the range of k values to ry in the classifier. Defaults to [1,200)
"""
def recommend(user_id, beer, brewery, k = range(1,200)):
    # get beer data
    df = sqlDataFrame("""SELECT * FROM `beers` LEFT JOIN breweries ON breweries.id = brewery_id
        WHERE beers.name LIKE %s AND breweries.name LIKE %s""", beer+'%', brewery + '%')
    # error handling
    if len(df) == 0:
        print "Error: user or beer/brewery not found."
        return
    
    # take first result
    df = df.ix[0,:]
    # get user data
    user_info = sqlDataFrame("SELECT lat, lng from users where id = %s", user_id)
    # calculate distance between user and brewery
    lat1, lng1 = df.lat, df.lng
    lat2, lng2 = user_info.lat[0], user_info.lng[0]
    df["distance"] = great_circle((lat1, lng1), (lat2, lng2)).km
    
    # now get training data - all of user's other ratings
    beers = sqlDataFrame("""SELECT ratings.*, beers.*, avg_rating FROM `ratings` LEFT JOIN beers ON (beer_id = beers.id) LEFT JOIN breweries
                    ON brewery_id = breweries.id
                     WHERE ratings.user_id = %s AND ratings.user_rating > 0 AND ratings.user_id IS NOT NULL""", user_id)
    # remove beers with no dat
    beers2 = beers[beers.ibu != 0]
    beers2 = beers2[beers2.abv != 0]

    # change ratings to integers
    Ytrain = beers2.user_rating*2

    # drop ratings and subset for features we want
    beers3 = beers2.drop("user_rating", 1)
    beers4 = beers3[["distance", "global_rating", "abv", "ibu", "style", "avg_rating"]]

    #discretize styles
    styles = {}
    count= 0
    for style in set(beers4.style):
        styles[style] = count
        count +=1
    beers5 = beers4.replace(styles)
    
    # then discretize style in input data
    df2 = df.replace(styles)
    # subset to match training data
    trimmed = df2[["distance", "global_rating", "abv", "ibu", "style", "avg_rating"]]
    input_beer = trimmed.as_matrix()
 
    # use all data to train
    Xtrain = beers5.as_matrix()
    
    # parameter tuning step
    params = {"n_estimators": k}
    classifier = sklearn.ensemble.RandomForestClassifier(n_estimators=k, random_state=42)
    grid_class = sklearn.grid_search.GridSearchCV(classifier, params, cv=10, scoring="f1")
    # extract and report results
    results = grid_class.fit(Xtrain,Ytrain)
    # set optimal tree number
    k_tuned = results.best_params_["n_estimators"]
    
    # now predict the input
    classifier = sklearn.ensemble.RandomForestClassifier(n_estimators=k_tuned, random_state=42)
    yhat = classifier.fit(Xtrain, Ytrain).predict(input_beer)
    
    # set cutoff to binarize
    cutoff = np.percentile(beers2.user_rating, 75)
    yhat2 = (yhat >= cutoff*2)
    
    print "Reccomendation: You %s like this beer. (Estimated rating: %.1f/5)" % ("will" if yhat2 else "won't", yhat/2)
```

```python
# lets try John Harvard's Harvard Square Helles for a user
recommend(46, "Harvard Square", "John Harvard's", range(1,5))
```

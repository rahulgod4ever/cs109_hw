```python
%matplotlib inline

from StringIO import StringIO
import numpy.linalg as lin
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import requests
import numpy as np
```

```python
exprs = pd.read_csv('HW2\\exprs_GSE5859.csv')
sampleinfo = pd.read_csv('HW2\\sampleinfo_GSE5859.csv')

exprs.sort(axis=1, inplace=True)
sampleinfo.sort(columns=['filename'], inplace=True)

print "Match between list of filenames in exprs and sampleinfo dfs: ",(sampleinfo['filename']==exprs.columns[:len(sampleinfo['filename'])]).all()
print exprs.head()
print sampleinfo.head()
```

```python
dates_arr = []
for date in sampleinfo['date']:
    dates_arr.append((pd.to_datetime(date, format = "%Y-%m-%d") - pd.to_datetime("2002-10-31",format="%Y-%m-%d")).days)

sampleinfo['elapsedInDays']=dates_arr
```

```python
#Only including samples with ethnicity CEU
sampleinfoCEU = sampleinfo[sampleinfo['ethnicity']=='CEU']
exprsCEU = exprs[exprs.columns[sampleinfo['ethnicity']=='CEU']]
```

```python
exprs = exprs[exprs.columns[sampleinfo['ethnicity']!='']]

avg_gene_intensity_CEU = np.mean([np.mean(exprsCEU[colname]) for colname in exprsCEU.columns])

avg_gene_intensity = np.mean([np.mean(exprs[colname]) for colname in exprs.columns])

for colname in exprsCEU.columns:
    exprsCEU[colname]-=[avg_gene_intensity_CEU]*len(exprsCEU[colname])

for colname in exprs.columns:
    exprs[colname]-=[avg_gene_intensity]*len(exprs[colname])
    
print exprsCEU.head()
print "Match between list of filenames in exprsCEU and sampleinfoCEU dfs: ",(sampleinfoCEU['filename']==exprsCEU.columns[:len(sampleinfoCEU['filename'])]).all()
print "Match between list of filenames in exprs and sampleinfo dfs: ",(sampleinfo['filename']==exprs.columns[:len(sampleinfo['filename'])]).all()

u,s,v = lin.svd(exprs, full_matrices=True)

```

```python
import scipy.linalg as scipylin
from sklearn import decomposition

S = scipylin.diagsvd(s,exprs.shape[0],exprs.shape[1])

np.allclose(np.dot(u,np.dot(S,v)),exprs)

svd = decomposition.TruncatedSVD(n_components=1)
x_1d = svd.fit_transform(exprs.T)

svd2 = decomposition.TruncatedSVD(n_components=2)
x_2d = svd2.fit_transform(exprs.T)
```

```python
print x_1d.shape

#Create colour list
c_eth = {}
list_eth = list(set(sampleinfo['ethnicity']))
for i in range(0,len(list_eth)):
    c_eth.update({list_eth[i]: i})

c = []
for eth in sampleinfo['ethnicity']:
    c.append(c_eth[eth])

```

```python
plt.scatter(x_1d,sampleinfo['elapsedInDays'], c = c)
```

```python
plt.scatter(x_2d[:,0],x_2d[:,1],c=c)
plt.axis((170,205,-30,40))
```

```python
from sklearn import cross_validation as cv
from sklearn import neighbors as neigh
from sklearn import metrics

def get_score(k):
    x = exprs.T
    x_train, x_test, y_train, y_test = cv.train_test_split(x,c,test_size = 0.5)

    knn = neigh.KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)

    y_pred = knn.predict(x_test)

    return metrics.accuracy_score(y_test,y_pred,normalize=True)

scores = []

k_max = 100

for i in range(1,k_max):
    scores.append(get_score(1))
print(np.mean(scores))
```
